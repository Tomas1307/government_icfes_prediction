```python
df_1=pd.read_csv('./data/datos_limpios.csv')
```

Estadisticas Descriptivas

```python
# Estadísticas generales para puntajes
df_1[["punt_matematicas", "punt_ingles", "eco"]].describe()
```

Histograma - Este comportamiento sugiere que, aunque la mayoría de los estudiantes tienen un desempeño medio en matemáticas, existe una dispersión considerable hacia ambos extremos. Esto podría estar relacionado con factores socioeconómicos, acceso a educación, o características individuales, lo que deberá analizarse en relación con otras variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma de puntajes de matemáticas
plt.figure(figsize=(8, 4))
sns.histplot(df_1["punt_matematicas"], bins=30, kde=True)
plt.title("Distribución puntajes Matemáticas")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show()
```

BoxPlot de Puntaje de matematicas: El boxplot muestra que la mayoría de los puntajes en matemáticas se concentran entre aproximadamente 40 y 70 puntos, con una mediana cercana a 50. También se observan múltiples valores atípicos tanto en los extremos bajos como altos, lo que indica la presencia de estudiantes con desempeños significativamente diferentes al promedio.

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_1["punt_matematicas"])
plt.title("Boxplot de Puntajes en Matemáticas")
plt.show()
```

Distribución de puntaje de matemáticas por género: La gráfica de violín muestra la distribución de los puntajes en matemáticas desagregados por género. Se observa que tanto hombres como mujeres presentan distribuciones similares, con medianas cercanas, aunque los hombres tienden a concentrarse ligeramente en puntajes más altos. Además, ambos géneros presentan colas largas que indican la presencia de valores extremos.

```python
plt.figure(figsize=(8, 4))
sns.violinplot(x="estu_genero", y="punt_matematicas", data=df_1)
plt.title("Distribución de Puntajes por Género (Violín)")
plt.xlabel("Género")
plt.ylabel("Puntaje Matemáticas")
plt.show()
```

Dispersión de puntaje de los estudiantes en pruebas de ingles y matematicas: La gráfica de dispersión entre los puntajes de matemáticas y ciencias naturales revela una relación positiva: a mayor puntaje en matemáticas, tiende a haber un mayor puntaje en ciencias naturales. La nube de puntos indica una alta densidad en puntajes medios (entre 40 y 70), y no se observan diferencias significativas en la relación entre géneros, ya que tanto hombres como mujeres están distribuidos de forma similar.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="punt_matematicas", y="punt_ingles", hue="estu_genero", data=df_1)
plt.title("Dispersión entre Matemáticas e Inglés")
plt.xlabel("Puntaje Matemáticas")
plt.ylabel("Puntaje Ciencias Naturales")
plt.show()
```

Conteo de Categorias Por Puntaje de Matematicas: La gráfica de barras muestra el conteo de estudiantes según su categoría de desempeño en matemáticas. La mayoría se concentra en los niveles "Bajo" y "Medio", mientras que una proporción significativamente menor alcanza el nivel "Alto". Esto sugiere que, aunque hay una base amplia de estudiantes con desempeño aceptable, aún existe un reto importante en elevar a más estudiantes a niveles de excelencia.

```python
def categorizar_puntaje(puntaje):
    if puntaje >= 70:
        return "Alto"
    elif puntaje >= 50:
        return "Medio"
    else:
        return "Bajo"

df_1["categoria_matematicas"] = df_1["punt_matematicas"].apply(categorizar_puntaje)
```

```python
sns.countplot(x="categoria_matematicas", data=df_1)
plt.title("Conteo de Categorías de Desempeño en Matemáticas")
plt.xlabel("Categoría")
plt.ylabel("Cantidad")
plt.show()
```
